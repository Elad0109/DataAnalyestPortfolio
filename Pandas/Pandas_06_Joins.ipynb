```python
emp=pd.DataFrame({"empid":[1,2,3,4],
                   "empname":['Shimi','Moshe','Noa','Avi'],
                   "deptid":[10,10,20,0]})
```

```python
emp
```

```python
dep
```

```python
emp.merge(dep,left_on="deptid",right_on="depid")
```

```python
#The default is inner join, can use how = inner
emp.merge(dep,
          left_on="deptid",
          right_on="depid",
          how="inner")
```

```python
#left outer join -- all rows from emp
mp.merge(dep,
          left_on="deptid",
          right_on="depid",
          how="left")
```

```python
#right outer join -- all rows from dep
emp.merge(dep,
          left_on="deptid",
          right_on="depid",
          how="right")
```

```python
#full outer join - inner+left+right=outer
emp.merge(dep,
          left_on="deptid",
          right_on="depid",
          how="outer")
```

```python
emp.merge(dep,
          how="cross")
```
