# Pandas DataFrame agg() Method

#Pandas Series.agg() is used to pass a function or list of function to be applied on a series or even each element of series separately.

```python
pup.groupby("Country")[["income"]].mean()
```

```python
pup.groupby("Country")[["income","avg"]].mean()
```

```python
pup.groupby("Country")[["income"]].agg("mean")
```

```python
pup.groupby("Country")[["income"]].mean()
```

```python
pup.groupby("Country")[["income"]].count()
```

```python
pup.groupby("Country")[["income"]].median()

```

```python
pup.groupby("Country")[["income"]].agg(["mean","count","median"])
```

```python
pup.count()
```

```python
pup.agg('count')#.to_frame()
```

```python
pup.mean(numeric_only=True)
```

```python
pup.agg(['count','max'])
```

```python
pup[["income","rooms"]].agg(['count','max','mean'])
```

```python
pup.groupby("Country").agg(['count','max'])
```

```python
pup.groupby("Country")[["income","rooms"]].agg(['count','max','mean'])
```

```python
#Different aggrigation per column
```

```python
pup
```

```python
pup.groupby("Country").agg({'Name':'count','income':['sum','mean','max'],'Age':'max','rooms':['mean','max']})\
    .rename(columns={"Name": "CountName"})
```

```python
var =pup.groupby("Country").agg({'Name':'count','income':['sum','mean','max'],'Age':'max','rooms':['mean','max']})
```

```python
var.rename(columns={"Name": "CountName"})
```

```python
pup.groupby("Country").agg(
    count_name=('Name','count'),
    mean_income=('income','mean'),
    max_age = ('Age','max'))
```
