```python
# and &
mask1=titanic["sex"]=="female"
mask2=titanic.age<=20
titanic[mask1&mask2]
```

```python
titanic[(titanic["sex"]=="female")&(titanic.age<=20)]
```

```python
# or |
mask1=titanic["sex"]=="female"
mask2=titanic.age<=20
titanic[mask1|mask2]
```

```python
titanic[(titanic["sex"]=="female")|(titanic.age<=20)]
```

```python
mask1=titanic["sex"]=="female"
mask2=titanic.age<=20
mask3=titanic.age>=50
titanic[mask1&mask2|mask3]
```

```python
mask1=titanic["sex"]=="female"
mask2=titanic.age<=20
mask3=titanic.age>=50
titanic[mask1&(mask2|mask3)]
```

```python
titanic[(titanic["sex"]=="female")&((titanic.age<=20)|(titanic.age>=50))]
```

```python
mask1=titanic["sex"]=="female"
mask2=(titanic.age<=20)|(titanic.age>=50)
titanic[mask1&mask2]
```

```python
mask1=titanic["sex"]=="female"
mask2=titanic.age<=20
mask3=titanic.age>=50
titanic[mask1&~(mask2|mask3)]
```
