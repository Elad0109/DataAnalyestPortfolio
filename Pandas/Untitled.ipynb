```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
pup=pd.read_csv('data/pupils.csv')
```

```python
pup.groupby(["Age","gen"])["Height"].mean()
```

```python
pup.groupby(["Age","gen"])["Height"].mean().unstack()
```

```python
pup.groupby(["Age","gen"])["Height"].mean().unstack(level=0)
```

```python
pup.groupby(["Age","gen"])["Height"].mean().unstack(level=1)
```

```python
pup.groupby(["Age","gen"])["Height"].mean().unstack().plot()
```
